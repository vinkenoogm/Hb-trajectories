```python
plt.rcParams.update({"pgf.rcfonts": False})

colors = ['red', 'green', 'orange', 'blue', 'purple']
labels = ['Cluster A', 'Cluster B', 'Cluster C', 'Cluster D', 'Cluster E']

fig, ax = plt.subplots(2, 4, figsize=(15, 8), sharex=True, sharey=True)

for a in fig.axes:
    a.tick_params(axis='both', labelsize=8, pad=1)   
    a.grid(which='major')


# Women on the left
for index, row in cenf2.iterrows():
    ax[0, 0].plot(cenf2.loc[index][:286], color=colors[index], label=labels[index])
ax[0, 0].set_ylim([7, 11])
    
for index, row in cenf3.iterrows():
    ax[0, 1].plot(cenf3.loc[index][:286], color=colors[index], label=labels[index]) 
    
for index, row in cenf4.iterrows():
    ax[0, 2].plot(cenf4.loc[index][:286], color=colors[index], label=labels[index])
    
for index, row in cenf5.iterrows():
    ax[0, 3].plot(cenf5.loc[index][:286], color=colors[index], label=labels[index])

# Men on the right
for index, row in cenm2.iterrows():
    ax[1, 0].plot(cenm2.loc[index][:286], color=colors[index], label=labels[index])
    
for index, row in cenm3.iterrows():
    ax[1, 1].plot(cenm3.loc[index][:286], color=colors[index], label=labels[index]) 
    
for index, row in cenm4.iterrows():
    ax[1, 2].plot(cenm4.loc[index][:286], color=colors[index], label=labels[index])
    
for index, row in cenm5.iterrows():
    ax[1, 3].plot(cenm5.loc[index][:286], color=colors[index], label=labels[index])

ax[1, 3].legend(loc='lower right', bbox_to_anchor=(-3.65, -1.4, 4.7, 1), ncol=3, mode="expand", fontsize=9)

cols = ['Women', 'Men']
rows = ['k = {}'.format(k) for k in range(2, 6)]

for axis, row in zip(ax[0], rows):
    axis.set_title(row, fontsize=9)

for axis, col in zip(ax[:,0], cols):
    axis.annotate(col, xy=(0, 0.5), xytext=(-axis.yaxis.labelpad - 5, 0),
                  xycoords=axis.yaxis.label, textcoords='offset points',
                  ha='left', rotation=90, va='center')
    axis.set_ylabel('Hb (mmol/L)', fontsize=9)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.2, hspace=None)
fig.text(0.5, -0.05, 'Days since first donation', ha='center', fontsize=9)    

# plt.savefig('fig_clusters_local.pdf', bbox_inches="tight")
# plt.savefig('fig_clusters_local.pgf', bbox_inches="tight")
```
